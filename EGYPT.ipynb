```python
import pandas as pd
import re

# Create your DataFrame
excel_1 = r"C:\Users\akhil\Downloads\Imp_excel\combined_excel.xlsx"
df = pd.read_excel(excel_1)
# df = merged_df
#df = merged_df_test
# Function to find patterns

def find_patterns(text):
    # Define the pattern
    pattern = r'SMWW \d{4}\s?[\w-]*|ISO \d{4}-\d|\bISO \d{4}\b'
    
    # Find all substrings that match the pattern
    matches = re.findall(pattern, text)
    
    # Remove ":" followed by any characters, and remove leading or trailing whitespaces.
    matches = [re.sub(r':.*', '', match).strip() for match in matches]
    
    # Handle "-" case specifically
    matches = [re.sub(r'-(\w)$', r' \1', match) for match in matches]
    
    # Remove the optional space between the digits and the trailing character
    matches = [re.sub(r'\s(\w)$', r'\1', match) for match in matches]

    return ', '.join(matches)


# Apply the function to the column
# Fill NaN values with empty strings and convert all values to strings
df['Standard Specification'] = df['Standard Specification'].fillna('').astype(str)

# Then apply the function
df['Standard Specification method'] = df['Standard Specification'].apply(find_patterns)
df['Standard Specification method'] = df['Standard Specification method'].apply(lambda x: x.rstrip(' -'))

display(df)
```
